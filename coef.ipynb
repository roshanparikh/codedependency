```python
import pickle as pkl
import plotly.express as px
import os, sys
from os.path import join
import shap
```

Models of interest:
1. XGBoost - 30 day - all features
2. SVC RBF - labs - 30 days

```python
file_path = join("output", 'shap', 
                 'model_results_all_30', 
                 'XGB', 
                 'shap_model_results_all_30_XGB.pkl'
                 )
with open(file_path, 'rb') as f:
    res = pkl.load(f)
```

```python
file_path
```

```python
exp_obj = shap.Explanation(values=res['values'], 
                           base_values=res['base_values'], 
                           data=res['data'], feature_names=res['feature_names'], output_names=res['output_names'])
```

```python
shap.plots.beeswarm(exp_obj, max_display=20)
```

```python
shap.plots.bar(exp_obj)
```
